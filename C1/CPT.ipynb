### Гистограмма и теоретическая плотность распределения Гумбеля

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline

mu = 0.5
b = 2
gumb_r = sts.gumbel_r(mu,b)
sample = gumb_r.rvs(1000)
plt.hist(sample, normed=True)
plt.ylabel('Fraction of samples')
plt.xlabel('$x$')
x = np.linspace(-5,15,100)
pdf = gumb_r.pdf(x)
plt.plot(x, pdf, label='Theoretical pdf', alpha=0.5, linewidth = 4)
```

### Гистограммы распределений выборочных средних 1000 выборок разных размеров и расчет параметров соответствующих нормальных распределений

```python
mass = [5,10,50]
graph_number = 1
plt.figure(figsize=(18,9))

# Математическое ожидание распределения Гумбеля E = mu + b*y, где y - константа Эйлера-Маскерони
# Дисперсия распределения Гумбеля D = sqrt(b*pi)/6

E = mu + b*0.57721
D = ((b*math.pi) ** 2)/6
xmin = -2
xmax = 7

for n in mass:
    mean_test = []
    for i in range(0,999):
        gumb_test = gumb_r.rvs(n)
        mean_test.append(sum(gumb_test)/n)
        i == i+1
    plt.subplot(1,len(mass),graph_number)
    plt.axis([xmin, xmax, 0, 1.4])
    plt.hist(mean_test, normed=True)
    plt.ylabel('Fraction of means samples, n = ' + str(n))
    plt.xlabel('$x$')
    plt.legend()
    x2 = np.linspace(xmin,xmax,100)
    norm_r = sts.norm(loc = E, scale = (D/n) ** 0.5)
    pdf2 = norm_r.pdf(x2)
    plt.plot(x2, pdf2, label='Theoretical pdf', alpha=0.5, linewidth = 4)
    graph_number += 1
```

Очевидно, что с ростом размера выборок распределение их выборочных средних становится все более сконцентрированным относительно теоретического среднего случайной величины, а дисперсия уменьшается. Все три распределения очень похожи на нормальное.
