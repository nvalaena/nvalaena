```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Распределение Фишера (F-распределение)
<a href="https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%A4%D0%B8%D1%88%D0%B5%D1%80%D0%B0">Описание распределения на Wikipedia</a>

## В теории вероятностей распределение Фишера — это двухпараметрическое семейство абсолютно непрерывных распределений
### Распределение Фишера имеет обозначение  $F=(d_1,d_2)$  где $d_1$ и $d_2$ - это числа степеней свободы. 

## Математическое ожидание: $\frac{d_2}{(d_2-2)}$ если $d_2>2$

## Дисперсия: $\frac{2\cdot d_{2}^{2}(d_1+d_2-2)}{d_1\cdot (d_2-2)^2\cdot (d_2-4)}$ если $d_2>4$

### Зададим числа степеней свободы и вычислим теоретические значения основных параметров распределения:

```python
d1=29 #показатель d1
d2=18 #показатель d2

#математическое ожидание (среднее)
m=d2/(d2-2.0)
print "Математическое ожидание:"
print m

#дисперсия
d=(2.0*d2**2*(d1+d2-2))/(d1*(d2-2)**2*(d2-4)) 
print "Дисперсия:"
print d
```

Сгенерируем выборку объёма 1000 из распределения Фишера (с $d1=29$ и $d2=18$), гистограмму выборки и теоретическую плотность:

```python
fisher_rv=sts.f(d1,d2)
sample=fisher_rv.rvs(size=1000)
x = np.linspace(0,6,1000)
pdf = fisher_rv.pdf(x)
plt.plot(x, pdf, label='theoretical CDF')
plt.hist(sample,bins=30,normed='true')
plt.ylabel('f(d1,d2)')
plt.xlabel('x')
```

### Сгенерируем 1000 выборок размерностями n=5,10 и 50. Для каждой выборки зафиксируем среднее значение, сохраним его в массиве и сформируем гистограмму. На гистограмму наложим график нормального распределения с параметрами, вычисленными на первом шаге. От дисперсии возьмем квадратный корень.

```python
n=[5,10,50]
matr=np.empty([3,1000])




#для каждой размерности выполняем цикл
i=0
for count_n in n:
    j=0
#Формируем выборку из 1000 значений размерности n где n=5,10,50
    while j<1000:
        matr[i,j]=fisher_rv.rvs(size=count_n).mean()
        j=j+1
    fig, ax = plt.subplots()
    #Подготовка данных для графика нормального распределения
    norm_rv = sts.norm(m, math.sqrt(d/count_n))
    sample_norm=norm_rv.rvs(1000)
    x_norm = np.linspace(0,6,1000)
    pdf_norm = norm_rv.pdf(x_norm)
    ax.plot(x_norm, pdf_norm , label='norm PDF')
    ax.hist(matr[i], bins=10,normed='true', label="vyborka("+str(count_n)+")")
    plt.ylabel("f(d1,d2) n="+str(count_n))
    plt.xlabel('x')
    plt.show()
    i=i+1
```

## С ростом n распределение средних становится более кучным и менее скошенным
## С увеличением размера выборки точность нормальной аппроксимации возрастает.
