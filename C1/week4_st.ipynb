# Гамма распределение

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
from scipy.stats import norm
%matplotlib inline
```

параметры гамма распределения

```python
a = 2.
scale = 1.
```

 Гистограмма выборки размером 1000 и график теоретической плотности распределения

```python
x = np.linspace(gamma.ppf(0.001, a, 0, scale), gamma.ppf(0.999, a, 0,scale), 1000)
plt.plot(x, gamma.pdf(x, a, 0, scale), 'r-', lw=5, alpha=0.6) #график плотности вероятности

r = gamma.rvs(a, 0, scale, size=1000)
plt.hist(r, 30, density=True, histtype='stepfilled', alpha=0.2) #гистограмма плотности вероятности
plt.title("Плотность вероятности гамма распределение") 
plt.xlabel("x") # ось абсцисс
plt.ylabel("y") # ось ординат
plt.show()



```

# ЦПТ своими руками
Плотность вероятности распределения средних значений в гамма распределении 

```python
k = list(range(1000))
x = np.linspace(norm.ppf(0.001), norm.ppf(0.999), len(k))
loc_norm = a # теоретическое среднее гамма распределения


for n in [5,10,15]:
    meds=[]
    for j in k:
        meds.append(np.mean(gamma.rvs(a, size=n)))
    plt.hist(meds, 50, density=True, histtype='stepfilled', alpha=0.2, label='эмпирические данные')
    
    scale_norm = np.sqrt(a/ n)  # теоретическое среднеквадратичное отклонение гамма распределения
    x = np.linspace(norm.ppf(0.001, loc_norm, scale_norm), norm.ppf(0.999, loc_norm, scale_norm), len(k))
    plt.plot(x, norm.pdf(x, loc_norm, scale_norm), 'r-', lw=5, alpha=0.6, label='теоретическое нормальное распределение')
    plt.legend(loc='best', frameon=False)
    # легенда графика
    plt.title("Выборка объемом n =%d" % n) 
    plt.xlabel("x") # ось абсцисс
    plt.ylabel("y") # ось ординат
    plt.show()
```


При увеличении размера выборки n, распределение средних значений начинает сходиться к 2 (параметр ф), среднеквадратичное отклонение падает. Т.е. при увеличении n точность апроксимации растет.

При этом расчетное нормальное распределение более менее совпадает с гистограммой, согласно закону больших чисел.
